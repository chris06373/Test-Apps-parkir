```python
class User:
    def __init__(self, username, password):
        self.username = username
        self.__password = password  # private attribute
        self.balance = 0

    def check_password(self, password):
        return self.__password == password

class Register:
    def __init__(self):
        self.users = {}

    def register(self, username, password):
        if username in self.users:
            print("Username sudah digunakan.")
        else:
            self.users[username] = User(username, password)
            print("Registrasi berhasil!")

    def login(self, username, password):
        user = self.users.get(username)
        if user and user.check_password(password):
            print("Login berhasil!")
            return user
        else:
            print("Username atau password salah.")
            return None

```

```python
import time
from datetime import datetime

class Vehicle:
    def __init__(motor, plate_number):
        motor.plate_number = plate_number
        motor.entry_time = datetime.now()
        motor.exit_time = None

    def exit(motor):
        motor.exit_time = datetime.now()

    def get_duration(motor):
        if motor.exit_time:
            duration = motor.exit_time - motor.entry_time
            return duration.total_seconds() / 60  # menit
        else:
            return 0

    def calculate_fee(motor, rate_per_hour=2000):
        duration_minutes = motor.get_duration()
        hours = max(1, int(duration_minutes // 60) + (1 if duration_minutes % 60 > 0 else 0))
        return hours * rate_per_hour

class Mobil:
    def __init__(mobil, plate_number):
        mobil.plate_number = plate_number
        mobil.entry_time = datetime.now()
        mobil.exit_time = None

    def exit(mobil):
        mobil.exit_time = datetime.now()

    def get_duration(mobil):
        if mobil.exit_time:
            duration = mobil.exit_time - mobil.entry_time
            return duration.total_seconds() / 60  # menit
        else:
            return 0

    def calculate_fee(mobil, rate_per_hour=4000):
        duration_minutes = mobil.get_duration()
        hours = max(1, int(duration_minutes // 60) + (1 if duration_minutes % 60 > 0 else 0))
        return hours * rate_per_hour

class Bus:
    def __init__(bus, plate_number):
        bus.plate_number = plate_number
        bus.entry_time = datetime.now()
        bus.exit_time = None

    def exit(bus):
        bus.exit_time = datetime.now()

    def get_duration(bus):
        if bus.exit_time:
            duration = bus.exit_time - bus.entry_time
            return duration.total_seconds() / 60  # menit
        else:
            return 0

    def calculate_fee(bus, rate_per_hour=6000):
        duration_minutes = bus.get_duration()
        hours = max(1, int(duration_minutes // 60) + (1 if duration_minutes % 60 > 0 else 0))
        return hours * rate_per_hour


class ParkingLot:
    def __init__(motor):
        motor.vehicles = {}

    def is_full(motor):
        return len(motor.vehicles) >= motor.capacity

    def vehicle_entry(motor, plate_number):
        if motor.is_full():
            print(f"Parkiran penuh! Kapasitas: {motor.capacity}")
        elif plate_number in motor.vehicles:
            print(f"Kendaraan {plate_number} sudah parkir.")
        else:
            motor.vehicles[plate_number] = Vehicle(plate_number)
            print(f"Kendaraan {plate_number} berhasil masuk. Slot tersisa: {motor.capacity - len(motor.vehicles)}")

    def vehicle_exit(motor, plate_number):
        if plate_number in motor.vehicles:
            del motor.vehicles[plate_number]
            print(f"Kendaraan {plate_number} keluar. Slot tersedia: {motor.capacity - len(motor.vehicles)}")
        else:
            print(f"Kendaraan {plate_number} tidak ditemukan.")

    def show_status(motor):
        print(f"Jumlah kendaraan saat ini: {len(motor.vehicles)} / {motor.capacity}")
        print("Daftar kendaraan:")
        for plate in motor.vehicles:
            print(f"- {plate}")

    def vehicle_entry(motor, plate_number):
        if plate_number in motor.vehicles:
            print(f"Kendaraan {plate_number} sudah masuk.")
        else:
            motor.vehicles[plate_number] = Vehicle(plate_number)
            print(f"Kendaraan {plate_number} masuk pada {motor.vehicles[plate_number].entry_time.strftime('%H:%M:%S')}")

    def vehicle_exit(motor, plate_number):
        vehicle = motor.vehicles.get(plate_number)
        if vehicle:
            vehicle.exit()
            duration = vehicle.get_duration()
            fee = vehicle.calculate_fee()
            print(f"Kendaraan {plate_number} keluar pada {vehicle.exit_time.strftime('%H:%M:%S')}")
            print(f"Durasi parkir: {duration:.2f} menit")
            print(f"Biaya parkir: Rp {fee}")
            del motor.vehicles[plate_number]
        else:
            print("Kendaraan tidak ditemukan.")


    def __init__(mobil):
        mobil.vehicles = {}

    def is_full(mobil):
        return len(mobil.vehicles) >= mobil.capacity

    def vehicle_entry(mobil, plate_number):
        if mobil.is_full():
            print(f"Parkiran penuh! Kapasitas: {mobil.capacity}")
        elif plate_number in mobil.vehicles:
            print(f"Kendaraan {plate_number} sudah parkir.")
        else:
            mobil.vehicles[plate_number] = Vehicle(plate_number)
            print(f"Kendaraan {plate_number} berhasil masuk. Slot tersisa: {mobil.capacity - len(mobil.vehicles)}")

    def vehicle_exit(mobil, plate_number):
        if plate_number in mobil.vehicles:
            del mobil.vehicles[plate_number]
            print(f"Kendaraan {plate_number} keluar. Slot tersedia: {mobil.capacity - len(mobil.vehicles)}")
        else:
            print(f"Kendaraan {plate_number} tidak ditemukan.")

    def show_status(mobil):
        print(f"Jumlah kendaraan saat ini: {len(mobil.vehicles)} / {mobil.capacity}")
        print("Daftar kendaraan:")
        for plate in mobil.vehicles:
            print(f"- {plate}")

    def vehicle_entry(mobil, plate_number):
        if plate_number in mobil.vehicles:
            print(f"Kendaraan {plate_number} sudah masuk.")
        else:
            mobil.vehicles[plate_number] = Vehicle(plate_number)
            print(f"Kendaraan {plate_number} masuk pada {mobil.vehicles[plate_number].entry_time.strftime('%H:%M:%S')}")


    def vehicle_exit(mobil, plate_number):
        vehicle = mobil.vehicles.get(plate_number)
        if vehicle:
            vehicle.exit()
            duration = vehicle.get_duration()
            fee = vehicle.calculate_fee()
            print(f"Kendaraan {plate_number} keluar pada {vehicle.exit_time.strftime('%H:%M:%S')}")
            print(f"Durasi parkir: {duration:.2f} menit")
            print(f"Biaya parkir: Rp {fee}")
            del mobil.vehicles[plate_number]
        else:
            print("Kendaraan tidak ditemukan.")


    def __init__(bus):
        bus.vehicles = {}

    def is_full(bus):
        return len(bus.vehicles) >= bus.capacity

    def vehicle_entry(bus, plate_number):
        if bus.is_full():
            print(f"Parkiran penuh! Kapasitas: {bus.capacity}")
        elif plate_number in bus.vehicles:
            print(f"Kendaraan {plate_number} sudah parkir.")
        else:
            bus.vehicles[plate_number] = Vehicle(plate_number)
            print(f"Kendaraan {plate_number} berhasil masuk. Slot tersisa: {bus.capacity - len(bus.vehicles)}")

    def vehicle_exit(bus, plate_number):
        if plate_number in bus.vehicles:
            del bus.vehicles[plate_number]
            print(f"Kendaraan {plate_number} keluar. Slot tersedia: {bus.capacity - len(bus.vehicles)}")
        else:
            print(f"Kendaraan {plate_number} tidak ditemukan.")

    def show_status(bus):
        print(f"Jumlah kendaraan saat ini: {len(bus.vehicles)} / {bus.capacity}")
        print("Daftar kendaraan:")
        for plate in bus.vehicles:
            print(f"- {plate}")

    def vehicle_entry(bus, plate_number):
        if plate_number in bus.vehicles:
            print(f"Kendaraan {plate_number} sudah masuk.")
        else:
            bus.vehicles[plate_number] = Vehicle(plate_number)
            print(f"Kendaraan {plate_number} masuk pada {bus.vehicles[plate_number].entry_time.strftime('%H:%M:%S')}")


    def vehicle_exit(bus, plate_number):
        vehicle = bus.vehicles.get(plate_number)
        if vehicle:
            vehicle.exit()
            duration = vehicle.get_duration()
            fee = vehicle.calculate_fee()
            print(f"Kendaraan {plate_number} keluar pada {vehicle.exit_time.strftime('%H:%M:%S')}")
            print(f"Durasi parkir: {duration:.2f} menit")
            print(f"Biaya parkir: Rp {fee}")
            del bus.vehicles[plate_number]
        else:
            print("Kendaraan tidak ditemukan.")
```
